```python
file1 = open("billionLaughAttack.txt","r")
line = file1.readline()
ind = 0
startCounter = False
cpu_usage = []
time = []
while(line):
    line = line.strip()
    if startCounter:
        if( ind%9 == 0):
            cpu_usage.append(float(line))
            sec = (ind/9) * 10
            time.append(sec)
        ind += 1
    elif('-' in line):
        startCounter = True
    line = file1.readline()

print(time)
print(cpu_usage)
    
file1.close()
```
